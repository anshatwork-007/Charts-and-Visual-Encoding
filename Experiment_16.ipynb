```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data1 = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thur', 'Fri'],
    'Study_Hours': [2, 3, 4, 2.5, 5],
    'Marks': [50, 60, 55, 60, 70],
    'Attendence': [60, 70, 75, 80, 80],
    'Sleep_Hours': [6, 7, 6, 8, 5],
    'Assignment_Completed': [1,2,1,3,2]
}
df = pd.DataFrame(data1)
print(df)
```

```python
# Line Chart (Trend)
plt.plot(df['Days'], df['Study_Hours'], marker='s') # for markers squares = s, circle = o, star = *, triangle = ^
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Study Hours Trend')
plt.show()
```

```python
# Advanced Line chart
plt.figure(figsize=(10, 6))

plt.plot(df['Days'], df['Study_Hours'], marker='o', color = 'blue', label='Study Hours')
plt.plot(df['Days'], df['Marks'], marker='s', color = 'red', label='Marks')
plt.title('Study Hours vs Marks')
plt.xlabel('Days')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Bar char(Comparision)
plt.figure(figsize=(10, 6))
plt.bar(df['Days'], df['Marks'] ,color = 'turquoise')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('Marks Comparison')
plt.show()
```

```python
# Advanced Bar graph

bars = plt.bar(df['Days'], df['Marks'], color='turquoise')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y), ha='center', va='bottom')

plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid()
plt.show()
```

```python
# Histpgram (Distribution)
data2 ={
        'Marks':  [50,55,65,40,45,60,70,80,84,70,88,76,65,70,90,85,95]
}
df = pd.DataFrame(data2)
plt.hist(df['Marks'], bins = 5, edgecolor = 'black', alpha = 0.25)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks Distribution')
plt.show()
```

```python
# Scatter plot

import matplotlib.pyplot as plt

plt.scatter(df['Marks'], df['Study_Hours'], color = 'purple')
plt.ylabel('Marks')
plt.xlabel('Study Hours')
plt.title('Marks vs Study Hours')
plt.show()
```

```python
# Conditional scatter plot
# Example category
df['Result'] = ['Fail','Pass','Fail','Pass','Pass']
colors = ['red' if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'], df['Marks'], c=colors)

plt.title("Study Hours vs Marks (Result-wise)")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
# DataSet 2

import seaborn as sns
import pandas as pd

data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North','South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics',
                 'Food','Clothing','Electronics','Food']
}

df = pd.DataFrame(data)

print(df)
```

## Same Graph using Seaborn

```python
# Line plot
sns.lineplot(x='Day', y='Sales', data=df)
plt.title('Sales Trend')
plt.show()
```

```python
#Bar Plot
sns.barplot(x='Day', y='Profit', data=df)
plt.title("Seaborn Bar Plot")
plt.show()
```

```python
#Histogram
sns.histplot(df['Sales'], bins=5)
plt.title("Seaborn Histogram")
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Seaborn Scatter Plot")
plt.show()
```
